# BLENDING IMAGES

```
## addWeighted function (weighted sum) 

##formula = new_pixel = a*pixel1 + b*pixel2 + y

img1=cv2.imread('/content/pupp2.jpg')
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2=cv2.imread('/content/watermark.png')
```

```
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```
plt.imshow(img1)
```

```
plt.imshow(img2)
```

```
img1.shape,img2.shape
```

```
#resize the images so that they are equal to each other

img1=cv2.resize(img1,(1200,1200))
img2=cv2.resize(img2,(1200,1200))
```

```
blended = cv2.addWeighted(src1=img1,alpha=0.8,src2=img2,beta = 0.2, gamma=10)
```

```
plt.imshow(blended)
```

```
##BLEND TOGETHER IMAGES OF DIFFERENT SIZES
```

```
img1=cv2.imread('/content/pupp2.jpg')
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2=cv2.imread('/content/watermark.png')
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```
img2=cv2.resize(img2,(50,50))
```

```
plt.imshow(img1)
```

```
img1.shape #y axis=177, xaxis=284
```

```
x_offset = 284 - 50
y_offset = 177-50
```

```
img2.shape
```

```
rows,cols,chal = img2.shape
```

```
roi = img1[y_offset:177,x_offset:284]
```

```
plt.imshow(roi)
```

```
img2gray = cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)
```

```
plt.imshow(img2gray, cmap='gray')
```

```
mask_inv = cv2.bitwise_not(img2gray)
```

```
plt.imshow(mask_inv,cmap='gray')
```

```
mask_inv.shape
```

```
import numpy as np
white_bck = np.full(img2.shape,255,dtype=np.uint8)
```

```
white_bck.shape
```

```
bk = cv2.bitwise_or(white_bck,white_bck,mask=mask_inv)
```

```
plt.imshow(bk)
```

```
fg = cv2.bitwise_or(img2,img2,mask=mask_inv)
```

```
plt.imshow(fg)
```

```
final_roi=cv2.bitwise_or(roi,fg)
```

```
plt.imshow(final_roi)
```

```
large_img=img1
small_img = final_roi
```

```
large_img[y_offset:y_offset+small_img.shape[1],x_offset:x_offset+small_img.shape[0]]=small_img
```

```
plt.imshow(large_img)
```
