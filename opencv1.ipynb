```
#drawing on images
blank_img = np.zeros(shape=(512,512,3),dtype='int16')
```

```
blank_img.shape
```

# drawing rectangle

```
plt.imshow(blank_img) # black because everything is zero
```

```
#drawing a rectangle on image
cv2.rectangle(blank_img,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
```

```
plt.imshow(blank_img)
```

```
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(255,0,0),thickness = 10)
```

```
plt.imshow(blank_img)
```

#drawing circle

```
cv2.circle(blank_img,center=(100,100),radius=50,color=(255,0,255),thickness= 8)
plt.imshow(blank_img)
```

```
cv2.circle(blank_img,center=(400,400),radius=50,color=(255,0,255),thickness= -1)
plt.imshow(blank_img) #filled in with color
```

# drawing line

```
cv2.line(blank_img,pt1=(0,0),pt2=(512,512),color=(102,255,255),thickness=5)
plt.imshow(blank_img)
```

#writing text

```
#custokm teext
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text='hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness = 3, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

#drawing polygon

```
blank_img = np.zeros(shape=(512,512,3),dtype='int32')
plt.imshow(blank_img)
```

```
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype='int32')
```

```
vertices.shape #open cv rakes 3 dimensions we need to reshape it
```

```
pts = vertices.reshape((-1,1,2))
```

```
pts.shape
```

```
cv2.polylines(blank_img,[pts],isClosed=True,color=(255,0,0),thickness = 5)
plt.imshow(blank_img)
```
